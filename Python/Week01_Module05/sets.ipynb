## Sets in Python

```python
s = {1,2,3,4,4,1,2} # set with duplicate elements
print(type(s))
print(s)
```

```python
s1 = set()  # empty set
print(type(s1))
print(s1)
s1.add(5)
s1.add(10)
print(s1)
```

```python
s2 = {} # This creates an empty dictionary, not a set
print(type(s2))
```

```python
s3 = set([1,2,3,4,5,5,4,3])
s3.add(6)
s3.remove(3)
s3.discard(10)  # discard does not raise an error if the element is not found

print(s3)
```

```python
# Set operations
A = {1,2,3,4}
B = {3,4,5,6}
print("Union:", A | B)
print("Intersection:", A & B)
print("Difference A-B:", A - B)
print("Difference B-A:", B - A)
print("Symmetric Difference:", A ^ B) # elements in either A or B but not in both
```

```python
# set iteration and length
nums = (10, 20, 30, 40, 50)
for num in nums:
    print(num)
print(len(nums))
```

```python
# Membership testing
A = {1,2,3,4,5}
B = {5,6,7,8,9}
print(3 in A)  # True   
print(10 in A) # False
print(5 not in B) # False
print(10 not in B) # True
```

```python
# Set comprehension
squares = {x**2 for x in range(10)}
print(squares)
# Frozen set
fs = frozenset([1,2,3,4,5]) # immutable set
print(type(fs))
print(fs)
# fs.add(6) # This will raise an error since frozensets are immutable
print(3 in fs)  # True
print(10 in fs) # False
```
