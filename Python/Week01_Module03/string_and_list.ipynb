### String splitting, joining and formatted String

```python
prompt1 = "Python is a programming Language."
token1 = prompt1.split()
print(token1)

prompt2 = "Python-is-a-programming-Language."
token2 = prompt2.split('-')
print(token2)

sentence = "-".join(token1)
print(sentence)
```

```python
name = "Mehedy"
age = 29
height = 5.110003435

info = "Name: {}, Age: {}, Height: {:.2f} feet".format(name, age, height)
print(info)

info_f = f"Name: {name}, Age: {age}, Height: {height:.2f} feet"
print(info_f)
```
