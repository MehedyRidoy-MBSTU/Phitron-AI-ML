### List as Stack and Queue

Stack is a LIFO data structure (Last In First Out). 
New elements are added to the top of the stack and removed from the top of the stack.

append() -> push

pop() -> pop

```python
stack = []
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)
print(f"Top element: {stack[-1]}")
```

```python
queue = []
queue.append(10)
queue.append(20)
queue.append(30)
print(queue)
print(f"Front element: {queue[0]}")
front = queue.pop(0)
print(f"Removed element: {front}")
print(queue)
```
