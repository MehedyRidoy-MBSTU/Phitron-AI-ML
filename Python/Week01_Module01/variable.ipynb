```python
eq = 10 + 10/2 - 5*2
```

```python
eq
```

```python
x = input("Enter value of x: ")
y = input("Enter value of y: ")
print(x, y)  # concatenation
```

```python
x = int(x[-1])  # last digit
y = int(y[-1])  # last digit
print(x + y)  # addition
```

```python
last_digit_of_x = x % 10
last_digit_of_y = y % 10

print(last_digit_of_x + last_digit_of_y)
```
